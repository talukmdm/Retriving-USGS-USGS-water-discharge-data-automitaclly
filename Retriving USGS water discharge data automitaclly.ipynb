```python
import glob
f=glob.glob("Area\\**.txt",recursive=True)
dfx=pd.DataFrame(columns=["basin","area"],index=(range(len(f))))
for i in range(len(f)):
    file=open(f[i],"r")
    name=f[i][10:18]
    dfx["basin"][i]=name
    dfx["area"][i]=file.read()
dfx["area"]=pd.to_numeric(dfx.area)
```

```python
df=dfx.loc[0:10]
df
```

```python
import numpy as np
for j in range(len(df['basin'])):
    basin=df['basin'][j]
    herring=hf.NWIS(basin, 'dv','1948-01-01', '2020-10-31',parameterCd="00060")
    main=herring.df()
    
    colm = 'USGS:'+basin+':00060:00003'
    main=main[[colm]]
    main=main.rename(columns={main.columns[0]:"discharge(ft3/s)"})
    main["stream_m3s"]=0.0283168*main["discharge(ft3/s)"]
    main["streamflow"]=(main["stream_m3s"]/(dfx.iloc[j][1]*100))*864
    main["day"]=main.index.day
    main["month"]=main.index.month
    main['year']=main.index.year
    main['date']=main.index.date
    final=main[["stream_m3s","day","month","year","date","streamflow"]]
    final=final.reset_index(drop=True)
    final.to_csv("new_output/"+basin+".txt",index=False)
```
